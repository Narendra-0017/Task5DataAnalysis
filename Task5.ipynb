```python
import pandas as pd

# Create dummy sales data
data = {
    "Product": ["Laptop", "Laptop", "Mobile", "Mobile", "Tablet", "Tablet", "Headphones", "Headphones"],
    "Region": ["North", "South", "North", "East", "West", "South", "East", "West"],
    "Sales": [1200, 800, 1500, 900, 700, 600, 400, 500]
}

df = pd.DataFrame(data)

# Save as CSV
df.to_csv("sales.csv", index=False)

print("âœ… sales.csv file created!")
df
```

```python
import matplotlib.pyplot as plt

# Load CSV
df = pd.read_csv("sales.csv")

# Show first rows
print("First 5 rows of dataset:")
print(df.head())

# Group by Product
product_sales = df.groupby("Product")["Sales"].sum()
print("\nTotal sales per product:")
print(product_sales)

# Plot Product Sales
product_sales.plot(kind="bar", title="Total Sales per Product", color="skyblue")
plt.ylabel("Sales")
plt.xlabel("Product")
plt.show()

# Group by Region
region_sales = df.groupby("Region")["Sales"].sum()
print("\nTotal sales per region:")
print(region_sales)

# Plot Region Sales
region_sales.plot(kind="pie", autopct="%1.1f%%", title="Sales Distribution by Region")
plt.ylabel("")
plt.show()
```
